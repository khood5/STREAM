```python
plt.rcParams['figure.figsize'] = [100, 80]
matplotlib.rcParams.update({'font.size': 45})
# plt.subplot(1, 2, 1)
# (<Figure size 10000x4000 with 1 Axes>, <Axes: >)
fig, axs = plt.subplots(4,2)
figSamples = copy.deepcopy(smallDataSet)
fig.tight_layout(pad=2.0)
for row in axs:
    for col in row:
        data = figSamples.pop(0)
        y, sr = librosa.load(traingDataBasesPath+data[0])
        S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128, fmax=8000)
        S_dB = librosa.power_to_db(S, ref=np.max)
        img = librosa.display.specshow(S_dB, x_axis='time', y_axis='mel', sr=sr, fmax=8000, ax=col)
        fig.colorbar(img, ax=col, format='%+2.0f dB')
        col.set(title=f"Mel-frequency spectrogram {data[1]}")
plt.rcParams['figure.figsize'] = defultePlotSize
matplotlib.rcParams.update({'font.size': oringalFontSize})
```
